```python
import pandas as pd
import sqlite3 as sql
```

```python
sqladb = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None)
sqladb.shape
```

```python
sqladb["native-country"].unique()
```

```python
sqladb.columns =['age','workclass','fnlwgt','education','education-num','marital-status','occupation',
           'relationship','race','sex','capital-gain','capital-loss',
           'hours-per-week','native-country','salary_slab']
```

```python
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

## 1. Select 10 records from the adult sqladb

```python
pysqldf('SELECT * FROM sqladb LIMIT 10')
```

### 2. Show me the average hours per week of all men who are working in private sector

```python
query2 = """SELECT avg("hours-per-week") as Avg_Working_Hours FROM sqladb 
            WHERE workclass =' Private'
        """
pysqldf(query2)
```

### 3. Show me the frequency table for education, occupation and relationship, separately


```python
query3 = """
select education,occupation,relationship from sqladb
"""
pysqldf(query3)
```

## 4. Are there any people who are married, working in private sector and having a masters degree


```python
query4 = """ select * from sqladb where  
                "marital-status" like '%Married%'
                and workclass=' Private'
                and occupation is not null
                """
pysqldf(query4)
```

## 5. What is the average, minimum and maximum age group for people working in different sectors


```python
query5 = """
            select occupation, avg(age) as Average_Age, min(age) as Minimun_Age,max(age) as Maximun_Age
            from sqladb group by occupation
            """

pysqldf(query5)
```

## 6. Calculate age distribution by country


```python
query6 = """
            select "native-country"  as Country,age 
            from sqladb 
            group by "native-country"
        """
pysqldf(query6)
```

## 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
query7 = """
            select  "capital-gain" , "capital-loss" ,
            ("capital-gain" - "capital-loss") as "Net-Capital-Gain" 
            from sqladb
        """
pysqldf(query7)
```
